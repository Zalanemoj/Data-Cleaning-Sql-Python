### Price per unit Coloumn

```python
mask = (
    ((data['Price Per Unit'].isna()) | (data['Price Per Unit'] == 0)) & 
    (data['Quantity'] != 0)
)
```

```python
data.loc[mask, 'Price Per Unit'] = data.loc[mask, 'Total Spent'] / data.loc[mask, 'Quantity']
```

### Item Column

```python
data['Item'].isnull().sum()
```

```python
merged_data = pd.merge(
    data,
    product,
    left_on=['Category', 'Price Per Unit'],  
    right_on=['Beverages', '5.00'],         
    how='left'                              
)
```

```python
merged_data=merged_data.drop(columns=['Item','Beverages','5.00'])
```

```python
merged_data = merged_data.rename(columns={'Item_1_BEV': 'Item'})
```

```python
merged_data.head(10)
```

```python
merged_data['Status'] = np.where(merged_data['Discount Applied'].isnull(), 'Invalid', 'Valid')
```

```python
merged_data=merged_data.fillna('unknown')
```
